```python
def Cross(X,num):
    [length,feature] = X.shape
    for k in range(0,feature):
#         print(k)
        up_b = min(feature,k+num)
        for i in range(k+1,up_b):
            temp = np.zeros(length)
            a = np.array(X[:,i])
            b = np.array(X[:,i+k])
            temp = a*b
#             for j in range(0,length):
#                 temp[j] = X[j,i]*X[j,i+k]
            X = np.vstack([X.T,temp.T])
            X = X.T
    return X

def Cross3(X,num):
    [length,feature] = X.shape    
    X = Cross(X,num)
    for k in range(0,feature):
        print(k)
        up_b = min(feature,k+num)
        for m in range(k+1,up_b):
#             print(m)
            for n in range(m+1,up_b):
                temp = np.zeros(length)
                a = np.array(X[:,k])
                b = np.array(X[:,m])
                c = np.array(X[:,n])
                temp = a*b*c
                X = np.vstack([X.T,temp.T])
                X = X.T
    return X

def RMSE(Y,Y_Pred):
    N=len(Y)
    Err=0
    for i in range(N):
        Err=Err+(Y[i]-Y_Pred[i])**2
    Err=(1./N*Err)**(1./2.)
    return Err

def write_to_file(filename, predictions):
    with open(filename, "w") as f:
        f.write("Id,Prediction\n")
        for i,p in enumerate(predictions):
            f.write(str(i+1) + "," + str(p) + "\n")
```
